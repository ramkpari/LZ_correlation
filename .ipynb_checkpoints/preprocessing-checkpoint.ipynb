```python
# Library initialization 
import pandas as pd
import numpy as np 
import os



data = pd.read_csv('7_new_STLI_2020_Nov_23_1853.csv.')
```

```python
data['correct_answer'].str.strip('')
data['key_resp.keys'].str.strip('')
```

```python
# String comparison and allocation to correct_answer_verification

data.loc[data['correct_answer'] == data['key_resp.keys'],'correct_answer_verification'] = '1'
data.loc[data['correct_answer'] != data['key_resp.keys'],'correct_answer_verification'] = '0'
```
